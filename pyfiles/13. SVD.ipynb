## 13.2 SVD in Python

```python
A = np.random.randn(4,6)
#  Singular value return as vector (not matrix) 
# its SVD
U,s,Vt = np.linalg.svd(A)

# create Sigma from sigma's
S = np.zeros(np.shape(A))
np.fill_diagonal(S,s)


# show the matrices
_,axs = plt.subplots(1,4,figsize=(10,6))

axs[0].imshow(A,cmap='gray',aspect='equal')
axs[0].set_title('A\nThe matrix')

axs[1].imshow(U,cmap='gray',aspect='equal')
axs[1].set_title('U\n(left singular vects)')

axs[2].imshow(S,cmap='gray',aspect='equal')
axs[2].set_title('W\n(singular vals)')

axs[3].imshow(Vt,cmap='gray',aspect='equal')
axs[3].set_title('V\n(right singular vects)')

plt.tight_layout()
plt.show()
```
