### 1.2.3 Vector-scalar multiplication

```python
s = 2
a = [3, 4, 5] # list
b = np.array(a)
print(a*s)  # Repetition of list 
print(b*s)  # Vector-scalar ⭐️
```

### 1.2.5. Transpose

```python
# Row vector
r = np.array([ [1,2,3] ])

# orientationless array
l = np.array([1,2,3])
```

```python
# print out the vector, its transpose, and its double-transpose
print(r), print(' ')

# Transpose the row vector
print(r.T), print(' ')

# double-transpose
print(r.T.T)
     
```

```python
# Same for the orientationless array
print(l), print(' ')
print(l.T), print(' ')
print(l.T.T)
```

## 1.4 Vector-dot product
- 하나의 숫자로 두 벡터 사이의 관계를 나타낸다
- 두 벡터에서 대응되는 원소끼리 곱한 다음 모든 결과를 더한다
- 동일한 차원의 두 벡터에서만 성립한다

----------------------------------------------------------

- 내적은 두 벡터 사이의 유사성(similarity) 또는 매핑(mapping)의 척도로 해석

```python
v = np.array([1, 2, 3, 4])
w = np.array([5, 6, 7, 8])
np.dot(v,w)
```

```python
s = 10
np.dot(s*v, w)
```

### 1.4.1 The dot product is distributive

```python
# just some random vectors...
v = np.array([ 0,1,2 ])
w = np.array([ 3,5,8 ])
u = np.array([ 13,21,34 ])


# two ways to compute
res1 = np.dot( v, w+u )
res2 = np.dot( v, w) + np.dot(v, u)

res1, res2
```

## 1.5 other types of vector multiplication

### 1.5.1 Hardamard product

```python
import numpy  as np
a = np.array([5, 4, 8, 2])
b = np.array([1, 0, .5])
a* b
```

```python
# Hadamard product (원소별 곱)
# 각 가게별 매출 = (각 가게에서 판매한 수량) × (각 가게의 가격)
# 즉, 벡터끼리 원소별로 곱하면 → 가게별 매출 벡터가 나옴 
# 예: np.array([3, 5, 2]) * np.array([100, 200, 150]) = [300, 1000, 300] #vector

# Dot product (내적) 
# 전체 매출 = 모든 가게 매출을 합산한 값
# 예: [300, 1000, 300].sum() = 1600
# 즉, Hadamard 곱으로 나온 벡터의 원소를 더하면 내적이 됨  #scalar
```

## 1.5. Orthogonal decomposition

### Exercise1

```python
# The vectors
v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v+w


# now plot all three vectors
plt.figure(figsize=(6,6))

a1 = plt.arrow(0,0,v[0],v[1],head_width=.3,width=.1,color='k',length_includes_head=True)
a2 = plt.arrow(v[0],v[1],w[0],w[1],head_width=.3,width=.1,color=[.5,.5,.5],length_includes_head=True)
a3 = plt.arrow(0,0,vPlusW[0],vPlusW[1],head_width=.3,width=.1,color=[.8,.8,.8],length_includes_head=True)


# make the plot look a bit nicer
plt.grid(linestyle='--',linewidth=.5)
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend([a1,a2,a3],['v','w','v+w'])
plt.title('Vectors , , and ')
plt.savefig('Figure_01_02a.png',dpi=300) # write out the fig to a file
plt.show()
     
```

```python
# vector difference
vMinusW = v-w


# now plot all three vectors
plt.figure(figsize=(6,6))

a1 = plt.arrow(0,0,v[0],v[1],head_width=.3,width=.1,color='k',length_includes_head=True)
a2 = plt.arrow(0,0,w[0],w[1],head_width=.3,width=.1,color=[.5,.5,.5],length_includes_head=True)
a3 = plt.arrow(w[0],w[1],vMinusW[0],vMinusW[1],head_width=.3,width=.1,color=[.8,.8,.8],length_includes_head=True)


# make the plot look a bit nicer
plt.grid(linestyle='--',linewidth=.5)
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend([a1,a2,a3],['v','w','v-w'])
plt.title('Vectors , , and ')
plt.show()
```
